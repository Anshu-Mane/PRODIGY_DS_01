## Task 01: Data Visualization using Bar Chart and Histogram

**Objective:**
Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable.

```python
#Importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading the titanic dataset
titanic_df = pd.read_csv("data/titanic.csv")
```

```python
titanic_df.head()
titanic_df.info()
```

```python
#counting no.of male andd female passengers
gender_count = titanic_df['Sex'].value_counts()
print(gender_count)
```

```python
#creating a figure
plt.figure(figsize=(5,6))
```

```python
#ploting the bar chart
plt.bar(gender_count.index, gender_count.values, width=0.4, edgecolor='black')

#adding title and labels
plt.title('Gender Distribution of Titanic Passengers')
plt.xlabel('Gender')
plt.ylabel('No.of Passengers')

#Display the graph
plt.show()
```

```python
#Removing missing values
age = titanic_df['Age'].dropna()
```

```python
#Plotting Histogram for age distribution of passengers
plt.figure(figsize=(8,6))

plt.hist(age, bins=20, edgecolor='black')

#adding title and labels
plt.title('Age distribution of Titanic passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')

#Displaying the graph
plt.show()
```
